```python
from PIL import Image
from pylab import *
##image manipulation
import pandas as pd
from os import walk
import os
import numpy
dirpath="C:\\Users\\dee Jay\\Documents\\BAIM\\Unstructured\\Assignment 2\\image"
images=[]
image_files=os.listdir(dirpath)
for image in image_files:
    im = Image.open(dirpath+"\\"+image)
    im_rs = im.resize((100, 100))
    images.append(im_rs)
    imshow(im_rs)

```

```python
for image in images:
    figure()
    imshow(image)
```

```python
image_array=[]
for image in images:
    ##math representation
    im_m = array(image)
    im_grey_m =array(image.convert('L'))
    figure()
    imshow(im_grey_m)
    im_v = im_grey_m.flatten()
    print(im_m)
    print(im_grey_m)
    print(im_v)
    figure()
    hist(im_v,256, normed =True)
    image_array.append(im_grey_m)
```

```python
image_norm=[]
##normalization
for image in image_array:
    imhist, bins = histogram(image.flatten(), 256, normed=True)
    cdf = imhist.cumsum()
    cdf = 255 * cdf / cdf[-1]
    im2 = interp(image.flatten(), bins[:-1], cdf)
    im_norm = im2.reshape(image.shape)
    ##display
    figure(figsize=(30,8))
    subplot(1,4,1)
    imshow(image)
    subplot(1,4,2)
    imshow(im_norm)
    subplot(1,4,3)
    hist(im_grey_m.flatten(),256)
    subplot(1,4,4)
    hist(im2.flatten(),256)
    image_norm.append(im2.flatten())
```
